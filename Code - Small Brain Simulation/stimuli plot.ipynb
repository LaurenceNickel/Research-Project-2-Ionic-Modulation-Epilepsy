```python
import matplotlib.pyplot as plt

# Step 1: Read the data from the file
file_path = '/Users/regina/Library/CloudStorage/OneDrive-Persönlich/06 Project 2/SmallBrainSim/stimuli/sigmoid-1.0.txt'  # Replace with the actual path to your text file

# Step 2: Read and convert the data
with open(file_path, 'r') as file:
    entries = [float(line.strip()) for line in file]  # Directly convert to float

# print(entries)

# Step 3: Generate x-values (entry numbers)
x_values = range(1, len(entries) + 1)

# print(x_values)

# Step 4: Plot the data
plt.figure(figsize=(10, 6))
plt.plot(x_values, entries, linestyle='-')
plt.title('Stimuli Over Time')
plt.xlabel('time')
plt.ylabel('stimulus')
plt.grid(True)
plt.show()

from ipykernel import get_connection_file
import os

connection_file = get_connection_file()
notebook_directory = os.path.dirname(connection_file)
print(notebook_directory)
```
